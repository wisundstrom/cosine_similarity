# Metrics for comparing spatial distributions

### Single values to describe distribution

```python
def max_density(cluster_hist):
    """ The simplest approach is just to take the the density of the most popuated bin"""
    maxcount = cluster_hist.reshape(-1,1).max()
    density = maxcount/cluster_hist.reshape(-1,1).sum()
    return density
```

```python
max_densities = [max_density(histo) for histo in histograms]
max_densities
```

```python
binsize =.1
def hist2d_entropy(cluster_hist,binsize):
    """this calculates the shannon entropy of the 2d histogram.
    entropy is minimized at 0 when all events have probability 1 or 0, and maximized at log(number of events)
    when all events are equally likely"""
    
    clust_entropy = entropy(cluster_hist.reshape(-1,1)) - np.log(binsize)
    return clust_entropy[0]
```

```python
entropies = [hist2d_entropy(histo, binsize) for histo in histograms]
entropies
```

### 2 sample distribtuion comparisons

These can be used one we have some good candidates for baseline or otherwise non-fradulent distributions
They each take input as a list of distributions and compute pairwise scores, but could be made simpler by comparing
all distributions to a single baseline distribution.

```python
def pairwise_ks_array(cluster_dists, data="dist"):
    #this takes the empirical CDF for each distribtuion and finds the maximum difference
    
    if data == "dist":
        dist_array =[[ks_2samp(cluster1[0], cluster2[0])[0] for cluster1 in cluster_dists] for cluster2 in cluster_dists]
        
        
    elif data == "angle":
        dist_array =[[ks_2samp(cluster1[1], cluster2[1])[0] for cluster1 in cluster_dists] for cluster2 in cluster_dists]
                
    else:
        raise Exception('not a valid data arg')
        
    return dist_array
```

```python
def pairwise_ep_array(cluster_dists, data="dist"):
    # this compares the empirical characteristic function for each distribtuion (the fourier transform of the pdf)
    
    if data == "dist":
        dist_array =[[epps_singleton_2samp(cluster1[0], cluster2[0])[0] for cluster1 in cluster_dists] for cluster2 in cluster_dists]
        
        
    elif data == "angle":
        dist_array =[[epps_singleton_2samp(cluster1[1], cluster2[1])[0] for cluster1 in cluster_dists] for cluster2 in cluster_dists]
                
    else:
        raise Exception('not a valid data arg')
        
    return dist_array
```

## Spark agg functions testing for density, chisq and entropy

```python
sc = SparkSession.builder.getOrCreate()
sc.conf.set("spark.sql.execution.arrow.enabled", "true")
```

```python
spark_test_data = datasets.make_blobs(
    n_samples = 140000, centers = 7, n_features = 2, cluster_std = [20,200,20,200,200,200,200], shuffle = False,
    random_state = 15, center_box=(500,3500))

df_agi = pd.DataFrame(spark_test_data[0])
df_agi.columns = ['AGI','Refund']
df_agi['label'] = spark_test_data[1]
df_agi['id'] = df_agi.index
df_agi["AGI"] = df_agi["AGI"] * 20

SAMPLE_AMOUNT = 5000

ssamp_clust_01 = df_agi.sample(SAMPLE_AMOUNT)
ssamp_clust_02 = df_agi[((df_agi["label"] == 0) | (df_agi["label"] == 1))].sample(SAMPLE_AMOUNT)
ssamp_clust_03 = df_agi[((df_agi["label"] == 0) | (df_agi["label"] == 2))].sample(SAMPLE_AMOUNT)
ssamp_clust_04 = df_agi[((df_agi["label"] == 1) | (df_agi["label"] == 3) | (df_agi["label"] == 5))].sample(SAMPLE_AMOUNT)
ssamp_clusters = [ssamp_clust_01, ssamp_clust_02, ssamp_clust_03, ssamp_clust_04]
```

```python
for idx, cluster in enumerate(ssamp_clusters):
    cluster['cluster_label'] = idx
 
```

```python
all_sample_df = pd.concat(ssamp_clusters)
spark_df = sc.createDataFrame(all_sample_df)
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def max_density(xcol, ycol):
    
    xcol=(xcol-xcol.min())/(xcol.max()-xcol.min())
    ycol=(ycol-ycol.min())/(ycol.max()-ycol.min())
    
    num_bins=int(np.sqrt(len(xcol))+1)
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=num_bins)

    maxcount = cluster_hist.reshape(-1,1).max()
    
    density = maxcount/len(xcol)
    
    return density
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def entropy_2d(xcol, ycol):
    
    xcol=(xcol-xcol.min())/(xcol.max()-xcol.min())
    ycol=(ycol-ycol.min())/(ycol.max()-ycol.min())
    
    num_bins=int(np.sqrt(len(xcol))+1)
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=num_bins)
    
    binsize=xedges[1]-xedges[0]
    
    clust_entropy = entropy(cluster_hist.reshape(-1,1)) - np.log(binsize)
    
    return clust_entropy[0]
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def qr_chisq(xcol, ycol):
    
    xcol=(xcol-xcol.min())/(xcol.max()-xcol.min())
    ycol=(ycol-ycol.min())/(ycol.max()-ycol.min())
    
    sample = (zip(xcol,ycol))
    
    cluster_pattern = PointPattern(sample)
    shape= "rectangle"
    nx = 10
    ny = 10

    chisq = qs.QStatistic(cluster_pattern, shape = shape, nx = nx, ny = ny).chi2
    
    return chisq
    
```

```python
xcol = all_sample_df["AGI"]
ycol = all_sample_df['Refund']
xcol=(xcol-xcol.min())/(xcol.max()-xcol.min())
ycol=(ycol-ycol.min())/(ycol.max()-ycol.min())

sample = (zip(xcol,ycol))

cluster_pattern = PointPattern(sample)
shape= "rectangle"
nx = 10
ny = 10

chisq = qs.QStatistic(cluster_pattern, shape = shape, nx = nx, ny = ny).chi2
```

```python
cols = ['AGI','Refund']
spatial_agg = spark_df.groupby('cluster_label').agg(qr_chisq(*cols).alias('chisq'),
                                                    max_density(*cols).alias('density'),
                                                    entropy_2d(*cols).alias('entropy')
                                                   ).sort('cluster_label')
```
