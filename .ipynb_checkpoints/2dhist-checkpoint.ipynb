# Metrics for comparing spatial distributions

### Single values to describe distribution

```python
def max_density(cluster_hist):
    """ The simplest approach is just to take the the density of the most popuated bin"""
    maxcount = cluster_hist.reshape(-1,1).max()
    density = maxcount/cluster_hist.reshape(-1,1).sum()
    return density
```

```python
max_densities = [max_density(histo) for histo in histograms]
max_densities
```

```python
binsize =.1
def hist2d_entropy(cluster_hist,binsize):
    """this calculates the shannon entropy of the 2d histogram.
    entropy is minimized at 0 when all events have probability 1 or 0, and maximized at log(number of events)
    when all events are equally likely"""
    
    clust_entropy = entropy(cluster_hist.reshape(-1,1)) - np.log(binsize)
    return clust_entropy[0]
```

```python
entropies = [hist2d_entropy(histo, binsize) for histo in histograms]
entropies
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def max_density(xcol, ycol):
    
    xcol = (xcol-xcol.min()) / (xcol.max()-xcol.min())
    ycol = (ycol-ycol.min()) / (ycol.max()-ycol.min())
    
    num_bins=int(np.sqrt(len(xcol))+1)
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=num_bins)

    maxcount = cluster_hist.reshape(-1,1).max()
    
    density = maxcount/len(xcol)
    
    return density
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def entropy_2d(xcol, ycol):
    
    xcol = (xcol-xcol.min()) / (xcol.max()-xcol.min())
    ycol = (ycol-ycol.min()) / (ycol.max()-ycol.min())
    
    num_bins=int(np.sqrt(len(xcol))+1)
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=num_bins)
    
    binsize=xedges[1]-xedges[0]
    
    clust_entropy = entropy(cluster_hist.reshape(-1,1)) - np.log(binsize**2)
    
    return clust_entropy[0]
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def qr_chisq(xcol, ycol):
    
    xcol = (xcol-xcol.min()) / (xcol.max()-xcol.min())
    ycol = (ycol-ycol.min()) / (ycol.max()-ycol.min())
    
    sample = (zip(xcol,ycol))
    
    cluster_pattern = PointPattern(sample)
    shape= "rectangle"
    nx = int(np.sqrt(len(xcol))+1)
    ny = int(np.sqrt(len(xcol))+1)

    chisq = qs.QStatistic(cluster_pattern, shape = shape, nx = nx, ny = ny).chi2
    
    return chisq
    
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def np_chisq(xcol, ycol):
    
    xcol = (xcol-xcol.min()) / (xcol.max()-xcol.min())
    ycol = (ycol-ycol.min()) / (ycol.max()-ycol.min())
    
    num_bins = int(np.sqrt(len(xcol))+1)
    total_bins = num_bins**2
    total_rows = len(xcol)
    expected_bin = total_rows/total_bins
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=num_bins)
    
    chisq = np.sum((((cluster_hist-expected_bin)**2) / expected_bin))
    
    return chisq
    
```

The results of any of these histogram related methods can be sensitive to the choice of bin-width. above, the bins are chosen so there are sqrt(n) bins in each direction. These methods below instead use an approximation to the bin width that minimizes mean integrated square error, $h_k = 3.5 \sigma_{k}N^{-1/(2+d)}$ for each dimension k.

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def max_density_bin(xcol, ycol):
    
    xcol = (xcol-xcol.min()) / (xcol.max()-xcol.min())
    ycol = (ycol-ycol.min()) / (ycol.max()-ycol.min())
    
    xstd = np.std(xcol)
    ystd = np.std(ycol)
    
    hx = 3.5*xstd*np.power(len(xcol), -1/4)
    hy = 3.5*ystd*np.power(len(ycol), -1/4)
    
    nx = int(1/hx+1)
    ny = int(1/hy+1)
    
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=[nx,ny])

    maxcount = cluster_hist.reshape(-1,1).max()
    
    density = maxcount/len(xcol)
    
    return density
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def entropy_2d_bin(xcol, ycol):
    
    xcol = (xcol-xcol.min()) / (xcol.max()-xcol.min())
    ycol = (ycol-ycol.min()) / (ycol.max()-ycol.min())
    
    xstd = np.std(xcol)
    ystd = np.std(ycol)
    
    hx = 3.5*xstd*np.power(len(xcol), -1/4)
    hy = 3.5*ystd*np.power(len(ycol), -1/4)
    
    nx = int(1/hx+1)
    ny = int(1/hy+1)
    
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=[nx,ny])
    
    clust_entropy = entropy(cluster_hist.reshape(-1,1)) - np.log(hx*hy)
    
    return clust_entropy[0]
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)

def np_chisq_bin(xcol, ycol):
    
    xcol = (xcol-xcol.min()) / (xcol.max()-xcol.min())
    ycol = (ycol-ycol.min()) / (ycol.max()-ycol.min())
    
    
    xstd = np.std(xcol)
    ystd = np.std(ycol)
    
    hx = 3.5*xstd*np.power(len(xcol), -1/4)
    hy = 3.5*ystd*np.power(len(ycol), -1/4)
    
    nx = int(1/hx+1)
    ny = int(1/hy+1)
    
    
    total_bins = nx*ny
    total_rows = len(xcol)
    expected_bin = total_rows/total_bins
    
    cluster_hist, xedges, yedges = np.histogram2d(xcol,ycol, bins=[nx,ny])
    
    chisq = np.sum((((cluster_hist-expected_bin)**2) / expected_bin))
    
    return chisq
    
```

If only one column is given, the distribution is compared to a uniform distribution (which can be replaced with a gaussian or some other baseline "good" distribution later. We also test 1 dimensional versions of the 2d tests from above.

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)
def ep_score(xcol):
    """this compares the empirical characteristic function for each distribtuion (the fourier transform of the pdf)"""
    ycol = None
    
    xcol = xcol/np.sum(xcol)
    xmin = xcol.min()
    xmax = xcol.max()
    num_samples = len(xcol)
    
    if not ycol:
        ycol = np.random.uniform(xmin,xmax, num_samples)
    else:
        ycol = ycol/np.sum(ycol)
        
    ep_score = epps_singleton_2samp(xcol,ycol)[0]
        
    return ep_score
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)
def energy_dist(xcol):
    """ This can be roughly thought of as the difference in area under the cdfs of the distributions
    https://en.wikipedia.org/wiki/Energy_distance"""
    ycol = None
    xmin = xcol.min()
    xmax = xcol.max()
    xcol = xcol/np.sum(xcol)
    
    num_samples = len(xcol)
    
    if ycol:
        ycol = ycol/np.sum(ycol)
    else:
        ycol = np.random.uniform(xmin,xmax, num_samples)
        
        
    ed = energy_distance(xcol, ycol)
        
    return ed
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)
def w_distance(xcol):
    """This can be thought of as the amount of work required to change one distribution into the other
    https://en.wikipedia.org/wiki/Wasserstein_metric"""
    ycol = None
    xmin = xcol.min()
    xmax = xcol.max()
    xcol = xcol/np.sum(xcol)
    
    num_samples = len(xcol)
    
    if not ycol:
        ycol = np.random.uniform(xmin,xmax, num_samples)
    else:
        ycol = ycol/np.sum(ycol)
        
    w_dist = wasserstein_distance(xcol,ycol)
        
    return w_dist
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)
def entropy_1d_bin(xcol):
    """If only one argument is given, this computes the entropy of a histogram of the data
    If ycol is set to be a distribution, this computes the Kullback-Leibler Divergence of the two samples"""
    ycol = None
    
    xstd = np.std(xcol)
    hx = 3.5*xstd*np.power(len(xcol), -1/3)
    nx = int((np.max(xcol)-np.min(xcol))/hx+1)
    total_rows = len(xcol)
    expected_bin = total_rows/nx
    
    x_hist, xedges= np.histogram(xcol, bins=nx)
    
    if ycol:
        y_hist, yedges = np.histogram(ycol, bins=nx)
        y_hist = y_hist.reshape(-1,1)
    else:
        y_hist = None


    entropy_1d = entropy(x_hist, y_hist) - np.log((xedges[1]-xedges[0])/np.sum(xcol))
    
    return entropy_1d
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)
def chisq_bin_1d(xcol):
    
    xstd = np.std(xcol)
    hx = 3.5*xstd*np.power(len(xcol), -1/3)
    nx = int((np.max(xcol)-np.min(xcol))/hx+1)
    total_rows = len(xcol)
    expected_bin = total_rows/nx
    
    hist, xedges= np.histogram(xcol, bins=nx)
    
    chisq = np.sum((((hist-expected_bin)**2) / expected_bin))
    
    return chisq
```

```python
@pandas_udf("float", PandasUDFType.GROUPED_AGG)
def chisq(xcol):
    
    chisq = np.sum((((xcol-np.average(xcol))**2) / np.average(xcol)))
    
    return chisq
```
