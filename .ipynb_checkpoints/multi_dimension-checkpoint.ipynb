# Statistical Depth Metrics
These are some ways that we can take multi-dimensional data and assign each point a "depth" where larger depth means more central. We then can create ranks for our data and perform statistical comparisons based on depth.

```python
def mahal_depth(samples, df):
    #input df of samples and distribution to base depth on
    #samples can either be a dataframe or array of n dimensional data points
    #or a single data point
    
    df = np.array(df)
    samples = np.array(samples)
    if samples.ndim==1:
        samples = samples.reshape(1,-1)
    centroid = np.average(df, axis = 0)
    cov_mat = np.cov(df.T)
    s_i = np.linalg.inv(cov_mat)
    depths = []
    
    for row in samples:    
        cent_dist = row - centroid
        mahal = 1/(1 + (cent_dist @ s_i @ cent_dist.T))
        depths.append(mahal)
    
    if len(depths)==1:
        depths==depths[0]
    return depths
```

```python
def mahal_shrinkage(samples, df):
    """input df of samples and distribution to base depth on samples can either be a
    dataframe or array of n dimensional data points, or a single data point
    The ledoit wolf sshrinkage operation transforms the covariance matrix in a way that 
    makes finding the inverse easier.
    """
    
    df = np.array(df)
    samples = np.array(samples)
    if samples.ndim==1:
        samples = samples.reshape(1,-1)
    centroid = np.average(df, axis = 0)
    shrunk = LedoitWolf()
    s_i = shrunk.fit(df).precision_
    depths = []
    
    for row in samples:    
        cent_dist = row - centroid
        mahal = 1/(1 + (cent_dist @ s_i @ cent_dist.T))
        depths.append(mahal)
    
    if len(depths)==1:
        depths==depths[0]
    return depths
```

```python
def mahal_mcd(samples, df):
    """input df of samples and distribution to base depth on samples can either be a
    dataframe or array of n dimensional data points, or a single data point
    This version uses the mcd instead of the covariance matrix, which helps with outliers, 
    but might cause problems with multi-modal data.
    """
    
    df = np.array(df)
    samples = np.array(samples)
    if samples.ndim==1:
        samples = samples.reshape(1,-1)
    centroid = np.average(df, axis = 0)
    mcd = MinCovDet()
    s_i = mcd.fit(df).precision_
    depths = []
    
    for row in samples:    
        cent_dist = row - centroid
        mahal = 1/(1 + (cent_dist @ s_i @ cent_dist.T))
        depths.append(mahal)
    
    if len(depths)==1:
        depths==depths[0]
    return depths
```

```python
def simple_depth(samples, df):
    #input df of samples and distribution to base depth on
    #samples can either be a dataframe or array of n dimensional data points
    #or a single data point
    
    df = np.array(df)
    samples = np.array(samples)
    if samples.ndim==1:
        samples = samples.reshape(1,-1)
    centroid = np.average(df, axis = 0)
    
    depths = []
    
    for row in samples:    
        cent_dist = row - centroid
        length = 1/(1+np.sqrt(np.sum(np.square(cent_dist))))
        depths.append(length)
        
    depths = []
    
    if len(depths)==1:
        depths==depths[0]
    return depths
```

```python
#this is extremely slow, probably not salvagable
def l1_depth(samples, df):
    #input df of samples and distribution to base depth on
    #samples can either be a dataframe or array of n dimensional data points
    #or a single data point
    
    df = np.array(df.sample(500))
    samples = np.array(samples)
    if samples.ndim==1:
        samples = samples.reshape(1,-1)
    depths = []
    
    for row1 in samples:
        dists = []
        
        for row2 in df:
            dist = np.sum(np.abs(row1-row2))
            dists.append(dist)
            
        depths.append(1 / (1 + np.average(dists)))
            
        
    
    if len(depths)==1:
        depths==depths[0]
    return depths
```

```python
def l2_depth(samples, df):
    #input df of samples and distribution to base depth on
    #samples can either be a dataframe or array of n dimensional data points
    #or a single data point
    
    df = np.array(df.sample(500))
    samples = np.array(samples)
    if samples.ndim==1:
        samples = samples.reshape(1,-1)
    depths = []
    
    for row1 in samples:
        dists = []
        
        for row2 in df:
            dist = np.sqrt(np.sum(np.square(row1-row2)))
            dists.append(dist)
            
        depths.append(1 / (1 + np.average(dists)))
            
        
    
    if len(depths)==1:
        depths==depths[0]
    return depths
```

```python
def l2_depth_aff_inv(samples, df):
    #input df of samples and distribution to base depth on
    #samples can either be a dataframe or array of n dimensional data points
    #or a single data point
    
    df = np.array(df)
    samples = np.array(samples)
    if samples.ndim==1:
        samples = samples.reshape(1,-1)
        
    shrunk = LedoitWolf()
    s_i = shrunk.fit(df).precision_
    
    depths = []
    
    for row1 in samples:
        dists = []
        
        for row2 in df:
            diff = row1-row2
            dist = diff @ s_i @ diff.T
            dists.append(dist)
            
        depths.append(1 / (1 + np.average(dists)))
            
        
    
    if len(depths)==1:
        depths==depths[0]
    return depths
```

## Tests using Depth
spearman r: basically just the correlation coefficients of the ranks or depths

modified kruskal wallace: basically anova on ranks





```python
#modified kruskal wallis
# should be ok with samples of different length
def mkw(sample1, sample2):
    # statistic indicating if the samples are similarly distributed or not
    sample1 = np.array(sample1)
    sample2 = np.array(sample2)
    n1 = sample1.shape[0]
    n2 = sample2.shape[0]
    n = n1+n2
    
    s1_depths = [mahal_depth(sample,sample1)for sample in (sample1,sample2)]
    s2_depths = [mahal_depth(sample,sample2)for sample in (sample1,sample2)]
    
    s1_ranks = rankdata((s1_depths[0]+s1_depths[1]))
    s2_ranks = rankdata((s2_depths[0]+s2_depths[1]))
    
    ranks_mat = np.array([s1_ranks[:n1], s1_ranks[n1:], s2_ranks[:n2], s2_ranks[n2:]])
    sum_sq_avg =[np.square(np.sum(ranks))/ranks.shape[0] for ranks in ranks_mat]
    
    
                       
    h_stat_1 = 6/(n*(n+1))
    h_stat_2 = np.sum(sum_sq_avg)
    h_stat_3 = 3*(n+1)
        
    return h_stat_1*h_stat_2-h_stat_3
```

```python
def baseline_kw(df, baseline, cols):
    """ This function takes the whole dataframe and a list of strings of the names of the numerical
    feature columns to be used and a distribution to use as the baseline for comparison"""
    
    columns = [df[x] for x in cols]
    sample2 = baseline
    
    @pandas_udf("float", PandasUDFType.GROUPED_AGG)
    def mkw(*columns):
        # statistic indicating if the samples are similarly distributed or not
        sample1 = np.array([*columns])
        
        n1 = sample1.shape[0]
        n2 = sample2.shape[0]
        n = n1+n2

        s1_depths = [mahal_depth(sample,sample1)for sample in (sample1,sample2)]
        s2_depths = [mahal_depth(sample,sample2)for sample in (sample1,sample2)]

        s1_ranks = rankdata((s1_depths[0]+s1_depths[1]))
        s2_ranks = rankdata((s2_depths[0]+s2_depths[1]))

        ranks_mat = np.array([s1_ranks[:n1], s1_ranks[n1:], s2_ranks[:n2], s2_ranks[n2:]])
        sum_sq_avg =[np.square(np.sum(ranks))/ranks.shape[0] for ranks in ranks_mat]



        h_stat_1 = 6/(n*(n+1))
        h_stat_2 = np.sum(sum_sq_avg)
        h_stat_3 = 3*(n+1)

        return h_stat_1*h_stat_2-h_stat_3
    
    return mkw(*columns)


```
